### Trailing Lambda

```kotlin
listOf(1,2,3).forEach ({n-> println(n)})
//Trailing lambda - last param is lambda, so move it outside
listOf(1,2,3).forEach (){n-> println(n)}

listOf(1,2,3).forEach {n-> println(n)}
```

```kotlin
val add ={a:Int,b:Int->a+b}
val sub = {a:Int,b:Int->a-b}
val muti = {a:Int,b:Int->a*b}
val dev = { a: Int, b: Int->a/b}

fun caculate(a:Int,b:Int, operation: (Int,Int)->Int):Int {
    return operation(a, b)
}

val result_add = caculate(5,10, add)
println(result_add)
val result_sub = caculate(5,10, sub)
println(result_sub)
val result_muti = caculate(5,10, muti)
println(result_muti)
val result_dev = caculate(5,10, dev)
println(result_dev)
```

### Collection

```kotlin
data class Student(val name: String ,val grade: Double , val city: String )

val students = listOf(
    Student("Sokha",  8.5, "Phnom Penh"),
    Student("Dara",   6.0, "Siem Reap"),
    Student("Bopha",  9.2, "Phnom Penh"),
    Student("Vichea", 7.5, "Kampot"),
    Student("Leakhena", 5.8, "Siem Reap"),

    )
```

### Map --- Transform Every Element

```kotlin
val names = students.map { it.name }
println(names)
// ["Sokha", "Dara", "Bopha", "Vichea", "Leakhena"]
```

### Task: Every student gets +0.5

```kotlin
val newstudents = students.map { it.copy(grade = it.grade + 0.5) }
println(newstudents)
```

### filter -> sortedBy -> groupBy -> fold

```kotlin
val passing = students.filter { it.grade >= 8.0}
println(passing)

val passing2 = students.filter { it.city == "Kampot"}
println(passing2)
```

### sortBy

```kotlin
val rankAscending = students.sortedBy { it.grade }
println(rankAscending)

val rankDescending = students.sortedByDescending { it.grade }
println(rankDescending)
```

### GroupBy

```kotlin
val byCity = students.groupBy { it.city }
println(byCity)// map(string,List<T>)
```

### Fold

```kotlin

val total = students.fold(0.0){acc,s ->acc + s.grade}

val topStudent = students.fold(students[0]){acc,s -> if (acc.grade > s.grade) acc else s}

println(total)
println(topStudent)
```
