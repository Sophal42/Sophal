## 1.when STATEMENT (switch)

```kotlin
val day= 3
when (day){
    1 ->println("Monday")
    2 ->println("Tuesday")
    3 ->println("Wednesday")
    else ->println("another day")
}
```

## 2. when EXPRESSION

```kotlin
val day = 5
val displayName = when(day) {
    1 ->"Monday"
    2 ->"Tuesday"
    3 ->"Wednesday"
    else ->"another day"
}
println(displayName)
```

### 3.when without argument

```kotlin
val tempurature =28
val advice = when {
    tempurature< 0 -> "Freezing -Stay inside"
    tempurature < 15 -> "cold - Bring a jacket"
    tempurature < 30 -> "comfortable"
    else -> "Hot - Stay hybrated"
}
println(advice)
```

### 4.Multiple values

```kotlin
val mouth = 61
val season = when(mouth) {
    12,1,2 -> "Winter"
    3,4,5 -> "Spring"
    6,7,8 -> "Summer"
    9,10,11 -> "Autumn"
    else -> "Invalid Input"
}
println(season)
```

## 5. Range checks inside when

```kotlin
val score = 90
val grade = when (score) {
    in 90..100 -> "A"
    in 75..89 -> "B"
    in 60..74 -> "C"
    else -> "F"

}
println(grade)
```
